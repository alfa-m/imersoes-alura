# Aula 3

## Usando pandas para manipular o dataframe original

### Cria um dataframe com as colunas selecionadas

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

### Renomeia as colunas

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo','Data':'data','Último (R$)':'valor_final','Var. Dia (%)':'var_dia_percent'}).copy()
df_principal
```

### Cria a coluna contendo a variação percentual e a coluna contendo o valor inicial

```python
df_principal['var_percent'] = df_principal['var_dia_percent'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (1+df_principal['var_percent'])
df_principal
```

### Cria correspodência sobre os dados dos tickers e renomeia a coluna de quantidade teórica

```python
df_principal = df_principal.merge(df_total_acoes, how='left', left_on='ativo', right_on='Código').copy()
df_principal = df_principal.drop(columns=['Código']).copy()
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

### Cria a coluna de variação em reais

```python
df_principal['var_rs'] = df_principal['qtd_teorica'] * (df_principal['valor_final'] - df_principal['valor_inicial'])
df_principal
```

### Formata os números para terem apenas 2 casas decimais e transforma a coluna "qtd_teorica" em número inteiro

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal['qtd_teorica'] = df_principal['qtd_teorica'].astype(int)
df_principal
```

### Cria a coluna informando o status da variação

```python
df_principal['resultado'] = df_principal['var_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável')).copy()
df_principal
```

### Cria correspodência sobre os dados dos tickers e seus nomes

```python
df_principal = df_principal.merge(df_ticker, how='left', left_on='ativo', right_on='Ticker').copy()
df_principal = df_principal.drop(columns=['Ticker']).copy()
df_principal = df_principal.rename(columns={'Nome':'nome'}).copy()
df_principal
```

### Cria correspodência sobre os dados das empresas e seus segmentos

```python
df_principal = df_principal.merge(df_chatgpt, how='left', left_on='nome', right_on='Nome da Empresa').copy()
df_principal = df_principal.drop(columns=['Nome da Empresa']).copy()
df_principal = df_principal.rename(columns={'Segmento':'segmento','Idade (aproximada)':'idade'}).copy()
df_principal
```

### Cria a coluna contendo a categoria de idades

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: 'Menor que 50 anos' if x < 50 else ('Maior que 100 anos' if x > 100 else 'Entre 50 e 100 anos')).copy()
df_principal
```

### Calcula constantes sobre a variação relevantes

Valores calculados:

*   Maior variação geral
*   Menor variação geral
*   Média da variação geral
*   Média da variação positiva
*   Média da variação negativa


```python
maior_var_rs = df_principal['var_rs'].max()
menor_var_rs = df_principal['var_rs'].min()
media_var_rs = df_principal['var_rs'].mean()
media_var_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_rs'].mean()
media_var_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_rs'].mean()

print(f'Maior variação: R$ {maior_var_rs:,.2f}')
print(f'Menor variação: R$ {menor_var_rs:,.2f}')
print(f'Média da variação: R$ {media_var_rs:,.2f}')
print(f'Média da variação positiva: R$ {media_var_subiu:,.2f}')
print(f'Média da variação negativa: R$ {media_var_desceu:,.2f}')
```

### Cria um dataframe contendo as variações positivas

```python
df_var_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_var_subiu.head()
```

### Cria um dataframe contendo as variações negativas

```python
df_var_desceu = df_principal[df_principal['resultado'] == 'Desceu']
df_var_desceu.head()
```

### Cria um dataframe contendo as variações por segmento

```python
df_analise_segmento = df_principal.groupby(by='segmento')['var_rs'].sum().reset_index()
df_analise_segmento.head()
```

### Cria um dataframe contendo as variações positivas por segmento

```python
df_analise_segmento_subiu = df_var_subiu.groupby(by='segmento')['var_rs'].sum().reset_index()
df_analise_segmento_subiu.head()
```

### Cria um dataframe contendo as variações negativas por segmento

```python
df_analise_segmento_desceu = df_var_desceu.groupby(by='segmento')['var_rs'].sum().reset_index()
df_analise_segmento_desceu.head()
```

### Cria um dataframe contendo as variações por status

```python
df_analise_saldo = df_principal.groupby(by='resultado')['var_rs'].sum().reset_index()
df_analise_saldo.head()
```

### Cria um dataframe contendo as variações por categoria de idade

```python
df_analise_idade = df_principal.groupby(by='cat_idade')['var_rs'].sum().reset_index()
df_analise_idade.head()
```

### Cria um dataframe contendo a quantidade de empresas por categoria de idade

```python
df_qtd_idade = df_principal.groupby(by='cat_idade')['var_rs'].count().reset_index()
df_qtd_idade = df_qtd_idade.rename(columns={'var_rs':'qtd'}).copy()
df_qtd_idade
```

## Criando gráficos

### Importa a biblioteca de criação de gráficos Plotly Express

```python
import plotly.express as px
```

### Cria um gráfico de barras contendo a variação em reais por status

```python
fig_barras_resultado = px.bar(data_frame=df_analise_saldo, x='resultado', y='var_rs',
                              text='var_rs',  text_auto='.2f',
                              title='Variação (R$) x Resultado')
fig_barras_resultado.show()
```

### Cria um gráfico de pizza contendo a variação em reais por segmento

```python
fig_pizza_segmento = px.pie(data_frame=df_analise_segmento_subiu, names='segmento',
                            values='var_rs', title='Variação por segmento')
fig_pizza_segmento.show()
```

### Cria um gráfico de barras contendo a variação em reais por categoria de idade

```python
fig_barras_cat_idade = px.bar(data_frame=df_analise_idade, x='cat_idade', y='var_rs',
                              text='var_rs',  text_auto='.2f',
                              title='Variação (R$) x Categoria de Idade')
fig_barras_cat_idade.show()
```

### Cria um gráfico de barras contendo a quantidade de empresas por categoria de idade

```python
fig_barras_idade = px.bar(data_frame=df_qtd_idade, x='cat_idade', y='qtd',
                              text='qtd',  text_auto='.0f',
                              title='Quantidade de Empresas x Categoria de Idade')
fig_barras_idade.show()
```
