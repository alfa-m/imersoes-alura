### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

### Cria um dataframe contendo os dados da performance das ações de janeiro de 2020 à dezembro de 2023

```python
dados_petr_original = yf.download('PETR4.SA', start='2020-01-01', end='2023-12-31')
dados_petr_original
```

### Cria cópia do dataframe com novos nomes de colunas

```python
dados_petr = dados_petr_original.copy()
dados_petr.columns = ['abertura', 'maximo', 'minimo', 'fechamento', 'fech_ajust', 'volume']
dados_petr
```

### Usa a coluna de data como índice

```python
dados_petr = dados_petr.rename_axis('data')
dados_petr
```

### Plota um gráfico de linha simples

```python
dados_petr['fechamento'].plot(figsize=(12,6))
plt.title('Variação do preço por data', fontsize=16)
plt.legend(['fechamento'])
```

### Cria um dataframe contendo os últimos 100 dados do dataframe original

```python
df_petr = dados_petr.tail(100).copy()
df_petr
```

### Adiciona a coluna de data

```python
df_petr['data'] = df_petr.index
df_petr
```

### Transforma os dados da coluna de data em números

```python
df_petr['data'] = df_petr['data'].apply(mdates.date2num)
df_petr
```

### Plota o dataframe usando Matplotlib

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Vamos definir a largura dos candles no gráfico
width = 0.7

for i in range(len(df_petr)):
    # Determinando a cor do candle
    # Se o preço de fechamento for maior que o de abertura, o candle é verde (a ação valorizou nesse dia).
    # Se for menor, o candle é vermelho (a ação desvalorizou).
    if df_petr['fechamento'].iloc[i] > df_petr['abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'

    # Desenhando a linha vertical do candle (mecha)
    # Essa linha mostra os preços máximo (topo da linha) e mínimo (base da linha) do dia.
    # Usamos `ax.plot` para desenhar uma linha vertical.
    # [df['Data'].iloc[i], df['Data'].iloc[i]] define o ponto x da linha (a data), e [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]] define a altura da linha.
    ax.plot([df_petr['data'].iloc[i], df_petr['data'].iloc[i]],
            [df_petr['minimo'].iloc[i], df_petr['maximo'].iloc[i]],
            color=color,
            linewidth=1)

    ax.add_patch(plt.Rectangle((df_petr['data'].iloc[i] - width/2,
                                min(df_petr['abertura'].iloc[i],
                                    df_petr['fechamento'].iloc[i])),
                               width,
                               abs(df_petr['fechamento'].iloc[i] - df_petr['abertura'].iloc[i]),
                               facecolor=color))

df_petr['media_movel_7'] = df_petr['fechamento'].rolling(window=7).mean()
df_petr['media_movel_14'] = df_petr['fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df_petr['data'], df_petr['media_movel_7'], color='orange', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df_petr['data'], df_petr['media_movel_14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias
# Adicionando legendas para as médias móveis
ax.legend()

# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotação para melhor legibilidade
ax.xaxis_date() #O método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando título e rótulos para os eixos x e y
plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adicionando uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

### Plota o dataframe usando Plotly

```python
# Criando subplots
'''
"Primeiro, criamos uma figura que conterá nossos gráficos usando make_subplots.
Isso nos permite ter múltiplos gráficos em uma única visualização.
Aqui, teremos dois subplots: um para o gráfico de candlestick e outro para o volume de transações."

'''
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

'''
"No gráfico de candlestick, cada candle representa um dia de negociação,
mostrando o preço de abertura, fechamento, máximo e mínimo. Vamos adicionar este gráfico à nossa figura."
'''
# Adicionando o gráfico de candlestick
fig.add_trace(go.Candlestick(x=df_petr.index,
                             open=df_petr['abertura'],
                             high=df_petr['maximo'],
                             low=df_petr['minimo'],
                             close=df_petr['fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# Adicionando as médias móveis
# Adicionamos também médias móveis ao mesmo subplot para análise de tendências
fig.add_trace(go.Scatter(x=df_petr.index,
                         y=df_petr['media_movel_7'],
                         mode='lines',
                         name='Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df_petr.index,
                         y=df_petr['media_movel_14'],
                         mode='lines',
                         name='Média Móvel 14 Dias'),
                         row=1, col=1)

# Adicionando o gráfico de barras para o volume
# Em seguida, criamos um gráfico de barras para o volume de transações, que nos dá uma ideia da atividade de negociação naquele dia
fig.add_trace(go.Bar(x=df_petr.index,
                     y=df_petr['volume'],
                     name='volume'),
                     row=2, col=1)

# Atualizando layout
#Finalmente, configuramos o layout da figura, ajustando títulos, formatos de eixo e outras configurações para tornar o gráfico claro e legível.
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider
                  width=1100, height=600)

# Mostrando o gráfico
fig.show()
```

## Importa a biblioteca Prophet

```python
from prophet import Prophet
```

## Cria um dataframe com dados das ações da Johnson & Johnson (JNJ)

```python
dados_jnj = yf.download("JNJ", start="2020-01-01", end="2023-12-31", progress=False)
dados_jnj = dados_jnj.reset_index()
```

## Cria o dataset de treino

```python
dados_jnj_treino = dados_jnj[dados_jnj['Date'] < '2023-07-31']
dados_jnj_treino
```

## Cria o dataset de teste

```python
dados_jnj_teste = dados_jnj[dados_jnj['Date'] >= '2023-07-31']
dados_jnj_teste
```

## Renomeia as colunas para os padrões da biblioteca Prophet

```python
dados_prophet_treino = dados_jnj_treino[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})
dados_prophet_treino
```

## Cria o modelo de previsão temporal

```python
modelo_prophet_jnj = Prophet(weekly_seasonality=True,
        yearly_seasonality=True,
        daily_seasonality=False)
modelo_prophet_jnj.add_country_holidays(country_name='US')
```

## Treina o modelo de previsão

```python
modelo_prophet_jnj.fit(dados_prophet_treino)
```

## Gera a previsão dos dados de ações

```python
futuro_jnj = modelo_prophet_jnj.make_future_dataframe(periods=150)
previsao_jnj = modelo_prophet_jnj.predict(futuro_jnj)
```

## Plota gráficos de linha contendo o conjunto de dados original e os dados previstos

```python
plt.figure(figsize=(14, 8))
plt.plot(dados_jnj_treino['Date'], dados_jnj_treino['Close'], label='Dados de Treino', color='blue')
plt.plot(dados_jnj_teste['Date'], dados_jnj_teste['Close'], label='Dados Reais (Teste)', color='green')
plt.plot(previsao_jnj['ds'], previsao_jnj['yhat'], label='Previsão', color='orange', linestyle='--')

plt.axvline(dados_jnj_treino['Date'].max(), color='red', linestyle='--', label='Início da Previsão')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Previsão de Preço de Fechamento vs Dados Reais')
plt.legend()
plt.show()
```
